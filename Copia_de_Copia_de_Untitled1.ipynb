```python
#Basado en el libro estadistica para administracion paguina 12
#Creamos una liat Vacia que presenta la edea de 100 personas
import random
edades=[]
#Hacemos un ciclo para generar 100 numeros aleatorios que representan la edad de las personas
for i in range(100):
  #Generamos un numero aleatorio con la funcion randint y lo guardamos en la variable numero aleatorio
  numero_aleatorio = random.randint(18,65)
  #Agregamos el numero aleatorio a la lista edades
  edades.append(numero_aleatorio)
  print(edades)
```

```python
import pandas as pd
```

```python
#Crea un DataFrame de pandas con los numeros aleatorios
df = pd.DataFrame(edades,columns=['Edad'])
df.head()
```

```python
#calcular la tabla de distribucion de frecuencias
tabla_frecuencias = df['Edad'].value_counts().reset_index()
tabla_frecuencias.columns = ['Edad','Frecuencia']
tabla_frecuencias = tabla_frecuencias.sort_values(by='Edad')

#imprimir la tabla de distribucion de frecuencias
print("Tabla de Distribucion de Frecuencias")
print(tabla_frecuencias)
```

```python
tabla_frecuencias['frec Porcentual'] = tabla_frecuencias['Frecuencia'] / 100 * 100
tabla_frecuencias
```

```python
tabla_frecuencias['frec Acumulada'] = tabla_frecuencias['Frecuencia'].cumsum()
tabla_frecuencias
```

Regla Studger
c= 1+log2(N)

```python
#importamos la libreria math de funciones matematicas
import math
```

```python
#volvemos a usar las listas de edades
print(edades)
```

```python
#crear un Dataframe con las edades
df = pd.DataFrame({'edad': edades})
df
```

```python
# calcular el numeros de intervalos utilizado la regla de Sturgues
n = len(edades)
K= int(1 + math.log2(n))
print("valor de n = ", n , "\nNumero de intervalos =", K)
```

```python
#crear tabla de frecuencia
tabla_frecuencia = pd.cut(df['edad'],bins=K).value_counts().sort_index()
tabla_frecuencia.columns = ['clase', 'Frec']
tabla_frecuencia
```

```python
#ordenar la tabla de frecuencias por el intervalo
tabla_frecuencia = tabla_frecuencia.sort_values(by="clase")
tabla_frecuencia
```

```python
# Genera columnas siguientes
tabla_frecuencia['frec Rel'] = tabla_frecuencia['frec'] / n
tabla_frecuencia['frec Porc'] = tabla_frecuencia['frec'] / n * 100
tabla_frecuencia['frec Ac'] = tabla_frecuencia['frec'].cumsum()
tabla_frecuencia['frec Rel Ac'] = tabla_frecuencia['frec Rel'].cumsum()
tabla_frecuencia['frec Port Ac'] = tabla_frecuencia['frec Porc'].cumsum()

#mostrar la tabla de frecuencia
print(tabla_frecuencia)
```
